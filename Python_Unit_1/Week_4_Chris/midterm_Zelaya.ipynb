## "Heatmap Analysis: Correlation Between CO2 Levels and Crop Yields"


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Organize the data
data = {
    'Year': ['1990s', '2020s', '2050s', '2080s'],
    'A1FI_CO2': [358, 432, 590, 810],
    'A2_CO2': [358, 432, 549, 709],
    'B1_CO2': [358, 421, 492, 527],
    'B2_CO2': [358, 422, 488, 561],
    'A1FI_Wheat_dY': [0, 4, 11, 18],
    'A2_Wheat_dY': [0, 4, 10, 18],
    'B1_Wheat_dY': [0, 3, 6, 8],
    'B2_Wheat_dY': [0, 3, 6, 11],
    'A1FI_Rice_dY': [0, 2, 10, 17],
    'A2_Rice_dY': [0, 2, 8, 17],
    'B1_Rice_dY': [0, 1, 5, 5],
    'B2_Rice_dY': [0, 1, 5, 10],
    'A1FI_Maize_dY': [0, 1, 4, 8],
    'A2_Maize_dY': [0, 1, 3, 7],
    'B1_Maize_dY': [0, 0, 1, 2],
    'B2_Maize_dY': [0, 0, 1, 4],
}

# Create a DataFrame
df = pd.DataFrame(data)

correlation_results = []

# Calculate correlation for each crop across all scenarios
for crop in ['Wheat', 'Rice', 'Maize']:
    for scenario in ['A1FI', 'A2', 'B1', 'B2']:
        correlation = df[[f'{scenario}_CO2', f'{scenario}_{crop}_dY']].corr().iloc[0, 1]
        correlation_results.append({'Crop': crop, 'Scenario': scenario, 'Correlation': correlation})

# Convert correlation results to a DataFrame
correlation_df = pd.DataFrame(correlation_results)

heatmap_data = correlation_df.pivot(index='Crop', columns='Scenario', values='Correlation')

#Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', center=0, linewidths=.5, fmt='.2f', cbar_kws={"shrink": .8})
plt.title('Heatmap of Correlation between CO2 Levels and Crop Yield Changes')
plt.xlabel('Scenario')
plt.ylabel('Crop')
plt.xticks(rotation=45)
plt.show()
```


**The heatmap illustrates the correlation between CO2 levels and the yields of three crops (Wheat, Rice, and Maize) across different scenarios:**

## Crop Analysis

### Wheat
- **Correlation**: Strong positive correlation with CO2 levels in all scenarios, especially in the high-emission A1FI scenario.
- **Implication**: Wheat yields are likely to increase with rising CO2, potentially benefiting from enhanced growth.

### Rice
- **Correlation**: Also shows strong positive correlations with CO2 across scenarios, particularly in A1FI.
- **Implication**: Higher CO2 levels may positively impact Rice yields, indicating a favorable response to increased carbon levels.

### Maize
- **Correlation**: Exhibits lower and more variable correlations with CO2 compared to Wheat and Rice.
- **Implication**: Maize may benefit from higher CO2, but other factors like temperature and water availability may limit its yield.

## Overall Insights
- **General Trend**: Increased CO2 levels generally lead to higher yields for Wheat and Rice, especially in high-emission scenarios.
